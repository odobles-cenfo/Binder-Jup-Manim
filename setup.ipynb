```python
# Cell 1:
from manim import *

# Cell 2:
%%manim -ql Formula
class Formula(Scene):
    def construct(self):
        # Example of a LaTeX formula rendered by Manim
        text = MathTex(
            "\\text{The derivative of } f(x) = x^2 \\text{ is }",
            "f'(x) = 2x"
        ).scale(1.2)
        
        self.play(Write(text))
        self.wait(2)
```
